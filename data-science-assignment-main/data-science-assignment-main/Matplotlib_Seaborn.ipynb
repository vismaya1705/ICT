```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
iris=sns.load_dataset("iris")
iris.head()
```

```python
sns.histplot(iris['sepal_length'],bins=10,kde=True)
plt.show()
```

```python
sns.pairplot(iris,hue="species")
plt.show()
```

```python
#violinplot

sns.violinplot(x='sepal_length',y='petal_width',data=iris)
plt.xticks(rotation=90)
plt.show()

```

```python
#swarmplot

sns.swarmplot(x='sepal_length', y='petal_width', data=iris,size=3)
plt.xticks(rotation=90)
plt.show()
```

```python
#stripplot

sns.stripplot(x='sepal_length', y='petal_width', data=iris, jitter=True)
plt.xticks(rotation=90)
plt.show()
```

```python
#boxplot

sns.boxplot(x='sepal_length', y='petal_width', data=iris)
plt.xticks(rotation=90)
plt.show()
```

```python
#barplot

sns.barplot(x='sepal_length', y='petal_width', data=iris)
plt.xticks(rotation=90)
plt.show()
```

```python
#countplot

sns.countplot(x='sepal_length', data=iris)
plt.xticks(rotation=90)
plt.show()
```

```python
iris_numeric = iris.drop(columns=['species'])
```

```python
corr = iris_numeric.corr()
```

```python
#heatmap

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```

```python
#barplot

sns.barplot(x='sepal_length', y='petal_width', data=iris)
plt.xticks(rotation=90)
plt.show()
```

```python
#scatterplot

sns.scatterplot(x='sepal_length', y='petal_width', data=iris)
plt.show()
```

```python
#line chart


sns.lineplot(x='sepal_length', y='petal_width', data=iris)
plt.show()
```
